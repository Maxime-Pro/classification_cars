```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cars_data['data'], cars_data['target'], random_state=0)
print("X_train shape : {}".format(X_train.shape))
print("y_train shape : {}".format(y_train.shape))
print("X_test shape : {}".format(X_test.shape))
print("y_test shape : {}".format(y_test.shape))
```

```python
cars_dataframe = pd.DataFrame(X_train, columns=cars_data['feature_names'])
pd.plotting.scatter_matrix(cars_dataframe, c=y_train, figsize=(15,15), marker='o', hist_kwds={'bins':20}, s=60, alpha=0.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
print("Test set predictions : \n {}".format(y_pred))
```

```python
print("Test set score : {:.3f}".format(np.mean(y_pred == y_test)))
print("Test set score : {:.3f}".format(knn.score(X_test, y_test)))
```
